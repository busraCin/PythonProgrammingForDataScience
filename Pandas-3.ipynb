```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)#çıktıda ki ... ifadesinden kurtulmamızı sağlar, gösterilecek max kolon sayısı olmasınhepsini göster 
df = sns.load_dataset("titanic")
df.head(4)
```

```python
#değişken veri seti içerisinde var mı
"age" in df
```

```python
#data frame olarak tip bilgisi alınması için çift [] parantez kullanılır
type(df[["age"]]) #tip bilgisi data frame
```

```python
type(df["age"]) #tek köşeli parantez kullanımında tip bilgisi pandas seri
```

```python
df["age2"] = df["age"] / 10  #yeni değişken ekleme varolan değişken üzerinden
df.head(3)
```

```python
#loc işlem yap, containes arama
df.loc[: , df.columns.str.contains("age")].head()
#içerisinde yaş ifadesini barındıran değişkenleri seçti
```

```python
df.loc[: , ~ df.columns.str.contains("age")].head(3)
# ~ ifadesi dışında anlamı taşır
```

```python
#iloc(integer based selection) tamsayı tabanlı seçim işlemleri 
df.iloc[0:3] # e kadar anlamı vardır
```

```python
#loc (label based selection) etiket tabanlı seçim işlemi, mutlak olarak seçer
df.loc[0:3] # e kadar anlamı yoktur son indexi de dahil eder.
```

```python
#iloc ve loc farkı iloc da seçimde string değer ile arama yapılmaz log da yapılır
df.loc[0:3 , "age"].head()
```

```python
#count ile ne kadar sorusuna cevep alırız
df[df["age"] > 50]["age"].count() 
```

```python
df.loc[df["age"] > 50 , ["age" , "class"]].head() 
#yaşı 50 den büyük olanların class bilgisi
```

```python
#yaşı 50 den byük ve erkek olanların class bilgisi, şart 2 oldu
df.loc[ (df["age"] > 50) & (df["sex"] == "male") , 
       ["age","class","survived"] ].head()
#birden fazla koşul yazılması durumunda her bir koşul parantez içine alınmalıdır
```

```python
df.loc[ (df["age"] > 50) & (df["sex"] == "male")
       & ((df["embarked"] == "C") | (df["embarked"] == "S")), #koşul içinde koşulda yine her bir koşul için ayrı parantez açılması gerekir
       ["age","class","survived","embarked"] ].head()
```
