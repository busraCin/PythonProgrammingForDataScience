```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = pd.read_csv("dataSets/breast_cancer.csv")
df = df.iloc[:, 1:-1] #veri setindeki problemli değişkenleri dışarıda bırakmak için kullanılır
df.head()
```

```python
#Korelasyon Analizi (Analysis of Correlation)
"""Korelasyon değişkenlerin birbiri ile ilişkisini ifade eden bir istatistiksel ölçümdür. 
-1 ile +1 arasında değerler alır.-1 e veya +1 e yaklaştıkça ilişkinin şiddeti kuvvetlenir.
2 değişkenin arasındaki ilişki pozitif ise buna 
pozitif korelasyon denir ve bir değişkenin değeri arttıkça diğerinin de değeri artar.
negatif korelasyonda birinin değeri artarken diğerinin azalır.
0 civari bir değer korelasyon olmadığı anlamına gelir. (-0.50) - 0 -(0.50) arası düşük korelasyon
2 değişken Yüksek korelasyon değerine sahip ise değişkenler aynı oldukları veya çok benzer özellik taşıdığı anlamına gelir
ve ikisinden birini bu sebeple çıkarmamız gerekir.
"""
num_cols = [col for col in df.columns if df[col].dtype in [int, float]]
corr = df[num_cols].corr() #korelasyon analizi hesaplamak için corr() fonksiyonu ullanılır.
corr
```

```python
sns.set(rc={'figure.figsize': (12, 12)}) #grafik 12 ye 12 lik olsun
sns.heatmap(corr, cmap="RdBu") #corr nesnesini yerleştirip ısı haritasına bakalım
plt.show()
#koyu maviler şiddetli pozitif korelasyonu, koyu kırmızılar kuvvetli negatif korelasyonu temsil eder.
```

```python
# Yüksek Korelasyonlu Değişkenlerin Silinmesi
cor_matrix = df.corr().abs() #mutlak değer olarak alma
```

```python
#aynı değerlerisilme, köşegen elemanları NaN olsun ve altındakiler silinsin.
upper_triangle_matrix = cor_matrix.where(np.triu #ilgili yapıya çevirme işlemi
                                         (np.ones(cor_matrix.shape), k=1) #1lerden oluşan ve oluşturduğumuz matrisin boyutunda nparrayi oluşturuyoruz
                                         .astype(bool))#np arrayini bool a çevir
upper_triangle_matrix
#çoklama, birbirini tekrar etme durumundan kurtarma
```

```python
#yüksek korelasyonlu değerleri silme için liste oluşturma
drop_list = [col for col in upper_triangle_matrix.columns 
             if any(upper_triangle_matrix[col]>0.90) ]
drop_list
```

```python
#df den silelim
cor_matrix[drop_list]
df.drop(drop_list, axis=1)
```

```python
#yukarıda ki işlemleri fonksiyonlaştıralım.
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={'figure.figsize': (15, 15)})
        sns.heatmap(corr, cmap="RdBu")
        plt.show()
    return drop_list


df.drop(drop_list, axis=1)
high_correlated_cols(df.drop(drop_list, axis=1), plot=True) #yeni korelasyon ısı grafiğimiz
```
