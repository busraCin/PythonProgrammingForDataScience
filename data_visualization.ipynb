```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
"""
kategorik değişen : sütun grafik, countplot(seaborn), bar;(matplotlib)
sayısal değişken: hist, bar
"""
```

```python
pd.set_option('display.max_columns' , None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()
```

```python
#value_counts() ilgili kategorik değişkeni betimler
df["sex"].value_counts().plot(kind='bar')
plt.show() #grafiği ekrana yazdır
```

```python
#sayısal değiken görselleştirme
plt.hist(df["age"]) #saısal değişkenin belirli aralıklara göre dağılımını verir
plt.show()
```

```python
plt.boxplot(df["fare"]) #veri setindeki aykırı değerleri çeyreklik değerler üzerinden gösterebilir
plt.show()
```

```python
#matplotlib özellikleri
#plot() x , y noktaları işaretlenerek görselleştirme elde edilir
x = np.array([1,13, 17, 18,20])
y = np.array([0,1,3,5,7])
plt.plot(x,y, 'o')
plt.show()
```

```python
#marker, işaretleyici
y = np.array([13,23,38,51,100])
plt.plot(y,marker="*")
plt.show()
```

```python
#line, çizgi gösterim
y = np.array([13,23,45,33,100])
plt.plot(y,linestyle="dotted" , color = "g")
plt.show()
```

```python
# Multiple Lines
x = np.array([23, 18, 31, 10])
y = np.array([13, 28, 11, 100])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
#LABELS
#Başlık
plt.title("Ana Başlık")
```

```python
#eksen isimlendirme
plt.xlabel("X ekseni isimlendirmesi")
plt.ylabel("Y ekseni isimlendirmesi")
plt.grid() #ızgara ifadesi ekleme
plt.show()
```

```python
# Subplots, birden fazla görselin çalıştırılması
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 2, 1) #1 satır, 2 sütun,1. grafiğini oluşturuyorum 
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 2, 2) #1 satır, 2 sütun, 2. grafiğini oluşturuyorum 
plt.title("2")
plt.plot(x, y)
plt.show()
```

```python
# 3 grafiği bir satır 3 sütun olarak konumlamak.
# plot 1
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 1)
plt.title("1")
plt.plot(x, y)

# plot 2
x = np.array([8, 8, 9, 9, 10, 15, 11, 15, 12, 15])
y = np.array([24, 20, 26, 27, 280, 29, 30, 30, 30, 30])
plt.subplot(1, 3, 2)
plt.title("2")
plt.plot(x, y)

# plot 3
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(1, 3, 3)
plt.title("3")
plt.plot(x, y)

plt.show()
```

```python
#seaborn
df = sns.load_dataset("tips")
df.head()
```

```python
#countplot kategorik değişken görselleştirme sns
df["sex"].value_counts()
sns.countplot(x =df["sex"], data=df) #cinsiyetin dağılımı, eksenler işaretlenmeli. x de kategori y de datanın alındığı dataframe adı
plt.show()
```

```python
#boxplot, Sayısal Değişken Görselleştirme sns
sns.boxplot(x=df["total_bill"])
plt.show()
```
