```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)
df = sns.load_dataset("titanic")
df.head()
```

```python
#apply satır veya sütunlarda otomatik olarak fonksiyon çalıştırmamızı sağlar
#lambda def gibi bir fonksiyon tanımlamamızı sağlar farkı kullan at özellikte olmasıdır.kod akışından sonra silinir
"""
for col in df.columns:
    if "age" in col:
        df[col] = df[col] /10
"""
df["age"].apply(lambda x: x/10).head()
```

```python
df.loc[: , df.columns.str.contains("age")].apply(lambda x: x/10).head()
```

```python
df.loc[: , df.columns.str.contains("age")].apply(lambda x: (x - x.mean() /  x.std())).head()
```

```python
def standart_scaler(col_name):
    return ((col_name - col_name.mean()) / col_name.std())
df.loc[: , ["age"]] = df.loc[: , df.columns.str.contains("age")].apply(standart_scaler)
df
```
