```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)#çıktıda ki ... ifadesinden kurtulmamızı sağlar, gösterilecek max kolon sayısı olmasınhepsini göster 
df = sns.load_dataset("titanic")
df.head()
```

```python
#kadınların ve erkeklerin ayrı ayrı yaş ortalamasına eriş
df.groupby("sex")["age"].mean()
#cinsiyete göre groupby işlemi ile yaş değikeninin ortalamasını almak problemi çözer
```

```python
#agg() 
df.groupby("sex").agg({"age" : "mean"})
```

```python
df.groupby("sex").agg({"age" : ["mean","sum","count"]})
#veriyi cinsiyet değişkenine göre kırdık. yaş değişeni üzerinde ayrı ayrı 3 işlem sonucunu aldık
```

```python
df.groupby("sex").agg({"age" : ["mean","sum"],
                     "survived" : "mean"}) 
#haytta kalma 0 ve 1 den oluşur ortalaması alınıca kadın ve erkeğin hayatta kalma yüzde ifadesine ulaşırız
```

```python
#iki seviyeli kırılma
df.groupby(["sex" , "embark_town"]).agg({"age" : ["mean" , "sum"], 
                                        "survived" : "mean"})
```

```python
df.groupby(["sex" , "embark_town","class"]).agg({"age" : ["mean" , "sum"], 
                                        "survived" : "mean",
                                                "sex" : "count"})
```
