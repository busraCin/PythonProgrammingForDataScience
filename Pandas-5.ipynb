```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)
df = sns.load_dataset("titanic")
df.head()
```

```python
#pivot table("kesişim","index","sütun"), 1. argümanın ön tanımlı değeri mean dir
#ilk argüman kesişimlerde, 2. argüman indexde(satır), 3.argüman sütunda hangi değişkeni görmek istediğimizi verir
df.pivot_table("survived" , "sex" , "embarked")
```

```python
#birden fazla boyut bilgisi ekleme
df.pivot_table("survived"  , ["embarked" , "class"] , "sex")
```

```python
""" cut ve qcut fonksiyonları
elimizde bulunan sayısal değişkenleri kategorik değişkenlere çevirmek için en yaygın kullanılan fonksiyondur.
Genelde , cut() sayısal değişkenleri hangi kategorilere böleceğimizi biliyorsak
qcut() elimizdeki sayısal değişkeni tanımıyorsak çeyreklik değerlerine göre bölünmesini istiyorsak kullanılır
"""
df["new_age"] = pd.cut(df["age"] , [0,10,18,25,40,90])
                          #neyi böleceğimiz, nerelerden böleceğimiz(sınıflandırma)
df.head()
```

```python
#şimdi yaş kırılımında hayatta kalmayı inceliyelim
df.pivot_table("survived" , "sex", "new_age")
```

```python
df.pivot_table("survived" , "sex", ["new_age" ,"class"])
```
